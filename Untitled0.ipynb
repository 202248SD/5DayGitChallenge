```python
import os

def convert_labels(input_label_dir, output_label_dir):
    os.makedirs(output_label_dir, exist_ok=True)

    for label_file in os.listdir(input_label_dir):
        input_file_path = os.path.join(input_label_dir, label_file)
        output_file_path = os.path.join(output_label_dir, label_file)

        with open(input_file_path, 'r') as infile, open(output_file_path, 'w') as outfile:
            for line in infile:
                parts = line.strip().split()

                center_x, center_y, width, height = parts[1:]
                outfile.write(f'0 {center_x} {center_y} {width} {height}\n')
```

```python
input_label_dir = '/content/Stanford-Dogs-Dataset-dog-breed-1/train/labels-m'
output_label_dir = '/content/Stanford-Dogs-Dataset-dog-breed-1/train/labels'
convert_labels(input_label_dir, output_label_dir)

input_label_dir = '/content/Stanford-Dogs-Dataset-dog-breed-1/valid/labels-m'
output_label_dir = '/content/Stanford-Dogs-Dataset-dog-breed-1/valid/labels'
convert_labels(input_label_dir, output_label_dir)

input_label_dir = '/content/Stanford-Dogs-Dataset-dog-breed-1/test/labels-m'
output_label_dir = '/content/Stanford-Dogs-Dataset-dog-breed-1/test/labels'
convert_labels(input_label_dir, output_label_dir)
```

```python
from google.colab import auth
from googleapiclient.http import MediaFileUpload
from googleapiclient.discovery import build

auth.authenticate_user()
```

```python
service = build('drive', 'v3')

def save_file_to_drive(name, path):
  file_metadata = {
    'name': name,
    'mimeType': 'application/octet-stream'
    }
  media = MediaFileUpload(path, mimetype='application/octet-stream',resumable=True)

  created = service.files().create(body=file_metadata,media_body=media,fields='id').execute()

  print('File ID: {}'.format(created.get('id')))

  return created
```
